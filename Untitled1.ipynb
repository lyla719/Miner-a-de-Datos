```python
# CONTEO DE ELEMENTOS
colores = ["amarillo", "azul", "morado", "rojo",
           "amarillo", "turquesa", "azul", "blanco"]

# Guardar los que están ya contados
contados = []

for color in colores:
    if color not in contados:
        print(color, "aparece", colores.count(color), "veces")
        contados.append(color)
#count=no.veces y append agregar elementos
```

```python
# LISTA DE ÚTILES INVERTIDA
utiles = ["lápiz", "cuaderno", "colores", "mochila"]

# lista original
print("Lista original:")
print(utiles)

# Invertir lista
utiles.reverse()
print("\nLista invertida:")
print(utiles)
```
